# Buildings and Building Units

## Example #1: Create a residential building

In this example, we will create a Project and a Branch, and add a Building with a Household to it.

#### Set things up

Required imports:

```python
from odeon.model import (
    Project,
    Branch,
    Building,
    BuildingType,
    RefurbishmentStatus,
    Household,
    Resident,
    Commercial,
    CommercialType,
    ScalingReference,
)
```

Create a project and a branch:

```python
my_project = Project()
my_branch = Branch(year=2023)
my_project.main_branch = my_branch
```

#### Create objects

Initialize an instance of a Building `my_building` and a Household `my_household` within, and add them to the Branch:

```python
my_building = Building()
my_building.usable_area = 140

my_household = Household()
my_household.net_floor_area = 120

my_building.add_building_units(my_household)

my_branch.add_objects(my_building)
```

#### Define building characteristics

The building can be enriched with a number of attributes:

```python
my_building.year_of_construction = 1969
my_building.building_type = BuildingType.DETACHED
my_building.refurbishment_status = RefurbishmentStatus.STANDARD_REFURBISHMENT
```

#### Add Residents

A household's size can either be controlled by the attributes `size`, or by adding a number of Residents:

```python
me = Resident()
my_friend = Resident()

my_household.add_residents([me, my_friend])
```

We can print the full object structure of our Project. Our branch by now contains everything we created, a default building geometry, and a device host.

```python
from odeon.processing.prints import print_project_stats, print_object_tree

print("Project Overview:\n")
print_project_stats(my_project)

print("\nBranch overview:\n")
print_object_tree(my_project.main_branch)
```

## Example #2: Create a Commercial Building

Analogously, commercial units can be created, for example in another empty building:

```python
my_work_place = Building()
my_work = Commercial()
my_work.commercial_type = CommercialType.OFFICE
my_work.scaling_reference = ScalingReference(name="employees", amount=65)

my_work_place.add_building_units(my_work)

my_branch.add_objects(my_work_place)
```

Our branch by now contains both buildings, one with a commercial unit and the other with a household:

```python
print("\nBranch overview after adding commercial building:\n")
print_object_tree(my_project.main_branch)
```
